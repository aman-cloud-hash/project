```python
import pandas as pd
import numpy as np
```

Load Dataset

```python
df = pd.read_csv("../data/uber_trips_dataset_50k.csv")

df.head()
```

```python
df.info()
```

```python
df.shape
```

Data Cleaning---Missing values remove

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

Convert Date Column

```python
df["pickup_time"] = pd.to_datetime(df["pickup_time"])
```

Time Features create karo

```python
df["Hour"] = df["pickup_time"].dt.hour
df["Day"] = df["pickup_time"].dt.day
df["Month"] = df["pickup_time"].dt.month
```

```python
df.head()
```

Daily Ride Distribution

```python
daily_rides = df.groupby("Day").size()

print(daily_rides)
```

Monthly Ride Distribution

```python
daily_rides = df.groupby(["Month","Day"]).size()

print(daily_rides)
```

```python
monthly_rides = df.groupby("Month").size()

print(monthly_rides)
```

Peak Hours Analysis

```python
peak_hours = df.groupby("Hour").size()

print(peak_hours)
```

```python
peak_hours.idxmax()
```

```python
peak_hours.max()
```

Busiest City / Zone

```python
df["city"].value_counts()
```

```python
df["city"].value_counts().idxmax()
```

Average Fare Analysis

```python
df["fare_amount"].mean()
```

Revenue Calculation

```python
df["Revenue"] = df["fare_amount"]
```

```python
df.head()
```

Revenue per City

```python
df.groupby("city")["Revenue"].sum()
```

```python
df.groupby("city")["Revenue"].sum().idxmax()
```

Outliers Check (NumPy)

```python
np.mean(df["fare_amount"])
```

```python
np.std(df["fare_amount"])
```

```python
np.max(df["fare_amount"])
```

Clean dataset save

```python
df.to_csv("../outputs/cleaned_uber_data.csv", index=False)
```
