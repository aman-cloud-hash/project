Data Cleaning

```python
df.isnull().sum()
```

You will see many missing values in:

Customer ID
Description

```python
df = df.dropna()
```

Explanation:

Missing customers = useless for analysis

Industry practice → remove them

```python
df = df.drop_duplicates()
```

Explanation:

Sometimes the same invoice is recorded twice.

```python
df = df[df["Quantity"] > 0]
```

Create Revenue Column

```python
df["Revenue"] = df["Quantity"] * df["Price"]
```

Example:

Quantity = 10
Price = 5

Revenue = 50

```python
df.head()
```

Time Features (Very Important)

Extract Year / Month

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

df["Year"] = df["InvoiceDate"].dt.year
df["Month"] = df["InvoiceDate"].dt.month
```

Companies analyze:

seasonal sales

monthly trends

yearly growth

```python
df.head()
```

Total Revenue

```python
total_revenue = df["Revenue"].sum()

print("Total Revenue:", total_revenue)
```

This tells the total sales of the company.

Monthly Sales Trend

```python
monthly_sales = df.groupby("Month")["Revenue"].sum()

print(monthly_sales)
```

Explanation:

groupby("Month")

Means:

Jan total sales

Feb total sales

Mar total sales


Best Selling Products

```python
top_products = df.groupby("Description")["Revenue"].sum().sort_values(ascending=False).head(10)

print(top_products)
```

This shows:

Top 10 products generating revenue

Companies use this for:

inventory planning

marketing

Worst Selling Products

```python
worst_products = df.groupby("Description")["Revenue"].sum().sort_values().head(10)

print(worst_products)
```

Companies remove these products.

Sales by Country

```python
country_sales = df.groupby("Country")["Revenue"].sum().sort_values(ascending=False)

print(country_sales)
```

Insight example:

UK = highest sales
France = second
Germany = third

Companies use this for market expansion.

Top Customers

```python
top_customers = df.groupby("Customer ID")["Revenue"].sum().sort_values(ascending=False).head(10)

print(top_customers)
```

This identifies:

VIP customers

Companies target them with discounts.

Average Order Value

```python
avg_order_value = df.groupby("Invoice")["Revenue"].sum().mean()

print("Average Order Value:", avg_order_value)
```

Very important business metric.
