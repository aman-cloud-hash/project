Date Conversion

```python
df["Date"] = pd.to_datetime(df["Date"])
```

Missing Values Clean

```python
df = df.fillna(0)
```

Country Filter

```python
india = df[df["Country/Region"] == "India"]
```

Time-wise Aggregation

```python
india_daily = india.groupby("Date")[["Confirmed","Deaths","Recovered","Active"]].sum()
```

```python
india_daily["New_Cases"] = india_daily["Confirmed"].diff()
```

Moving Average 7==("MA7")

```python
india_daily["MA7"] = np.convolve(
    india_daily["New_Cases"].fillna(0),
    np.ones(7)/7,
    mode="same"
)
```

Peak Analysis

```python
peak_cases = india_daily["New_Cases"].max()
peak_date = india_daily["New_Cases"].idxmax()

print(peak_cases)
print(peak_date)
```

Correlation

```python
correlation = india_daily["Confirmed"].corr(india_daily["Deaths"])

print(correlation)
```
